```python
import numpy as np
from scipy import stats
```

```python
# Task 1: Create Sample Data

data = [1.13, 1.55, 1.43, 0.92, 1.25,
        1.36, 1.32, 0.85, 1.07, 1.48,
        1.20, 1.33, 1.18, 1.22, 1.29]

print(data)
```

```python
# Task 2: Calculate Sample Mean and Sample Standard Deviation

sample_mean = np.mean(data)

sample_std = np.std(data, ddof=1)

n = len(data)

print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)
print("Sample Size:", n)
```

```python
# Task 3: Calculate 99% Confidence Interval Using Sample Standard Deviation

confidence_level = 0.99

t_value = stats.t.ppf((1 + confidence_level) / 2, df=n-1)

margin_error = t_value * (sample_std / np.sqrt(n))

lower_limit = sample_mean - margin_error
upper_limit = sample_mean + margin_error

print("99% Confidence Interval:")
print(lower_limit, "to", upper_limit)
```

```python
# Task 4: Confidence Interval Using Known Population Standard Deviation

population_std = 0.2

z_value = stats.norm.ppf((1 + confidence_level) / 2)

margin_error_known = z_value * (population_std / np.sqrt(n))

lower_limit_known = sample_mean - margin_error_known
upper_limit_known = sample_mean + margin_error_known

print("99% Confidence Interval with Known Population Standard Deviation:")
print(lower_limit_known, "to", upper_limit_known)
```

```python
# Task 5: Conclusion

print("The confidence interval estimates the range in which the true population mean is likely to fall.")
print("The t-distribution was used because the sample size is small and population standard deviation was initially unknown.")
```

```python
# Hypothesis Testing
# Task 1: Import Libraries
from scipy.stats import norm
```

```python
# Task 2: Define Given Values

sample_mean = 3050

mu = 1000 + (5 * 600)

sigma = 5 * 25

n = 25

alpha = 0.05

print("Theoretical Mean:", mu)
```

```python
# Task 3: State Hypotheses

print("Null Hypothesis (H0): Weekly operating cost is equal to theoretical cost.")

print("Alternative Hypothesis (H1): Weekly operating cost is higher than theoretical cost.")
```

```python
# Task 4: Calculate Test Statistic

z = (sample_mean - mu) / (sigma / np.sqrt(n))

print("Z Test Statistic:", z)
```

```python
# Task 5: Determine Critical Value

critical_value = norm.ppf(1 - alpha)

print("Critical Value:", critical_value)
```

```python
# Task 6: Decision Making

if z > critical_value:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")
```

```python
# Task 7: Conclusion

print("There is sufficient evidence to support the claim that weekly operating costs are higher than the theoretical model.")
```
