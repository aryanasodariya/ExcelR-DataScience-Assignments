```python
# Task 1: Import Libraries

import pandas as pd
import numpy as np
```

```python
# Task 2: Load Datasets

patient_df = pd.read_csv("Patient_Data.csv")
billing_df = pd.read_csv("Billing_Data.csv")
```

```python
# Task 3: Display Dataset Information

print(patient_df.info())
```

```python
# Task 4: Select Billing Related Columns

billing_columns = patient_df[['PatientID', 'Department', 'Doctor', 'BillAmount']]

print(billing_columns.head())
```

```python
# Task 5: Drop Administrative Columns

patient_df = patient_df.drop(columns=['ReceptionistID', 'CheckInTime'])

print(patient_df.head())
```

```python
# Task 6: Find Total Bill Amount Per Department

department_bill = patient_df.groupby('Department')['BillAmount'].sum()

print(department_bill)
```

```python
# Task 7: Remove Duplicate Patient Records

patient_df = patient_df.drop_duplicates(subset='PatientID')

print(patient_df.head())
```

```python
# Task 8: Fill Missing BillAmount Values

mean_bill = patient_df['BillAmount'].mean()

patient_df['BillAmount'] = patient_df['BillAmount'].fillna(mean_bill)

print(patient_df['BillAmount'])
```

```python
# Task 9: Merge Billing Dataset with Patient Dataset

merged_df = pd.merge(patient_df, billing_df, on='PatientID')

print(merged_df.head())
```

```python
# Task 10: Create New Patients DataFrame

new_patients = pd.DataFrame({
    'PatientID': [201, 202],
    'Department': ['Cardiology', 'Neurology'],
    'Doctor': ['Dr. Mehta', 'Dr. Shah'],
    'BillAmount': [4500, 5200]
})

print(new_patients)
```

```python
# Task 11: Concatenate New Patients Row-wise

updated_patient_df = pd.concat([patient_df, new_patients], ignore_index=True)

print(updated_patient_df.tail())
```

```python
# Task 12: Create New Billing Columns

new_columns = pd.DataFrame({
    'InsuranceCovered': ['Yes'] * len(updated_patient_df),
    'FinalAmount': updated_patient_df['BillAmount'] * 0.9
})

print(new_columns.head())
```

```python
# Task 13: Concatenate Billing Columns Column-wise

final_df = pd.concat([updated_patient_df, new_columns], axis=1)

print(final_df.head())
```

```python
# Task 14: Display Final Dataset Information

print(final_df.info())
print(final_df.head())
```
