```python
# Section:A
# Question 1

numbers = [10, 20, 30, 40, 50]

numbers.append(60)
print("Append:", numbers)

numbers.extend([70, 80])
print("Extend:", numbers)

numbers.insert(2, 25)
print("Insert:", numbers)

numbers.remove(40)
print("Remove:", numbers)

numbers.pop()
print("Pop:", numbers)

print("Index of 30:", numbers.index(30))

print("Count of 20:", numbers.count(20))

numbers.sort()
print("Sort:", numbers)

numbers.reverse()
print("Reverse:", numbers)

numbers.clear()
print("Clear:", numbers)
```

```python
# Section : C
# Question 11

import pandas as pd
import numpy as np

data = {
    "Name": ["Amit", "Rahul", "Priya", "Sneha", "Karan",
             "Riya", "Vikas", "Neha", "Aman", "Pooja"],
    
    "Age": [21, 22, 20, 23, 24,
            22, 25, 21, 23, 24],
    
    "City": ["Ahmedabad", "Surat", "Vadodara", "Rajkot", "Bhavnagar",
             "Ahmedabad", "Surat", "Rajkot", "Vadodara", "Bhavnagar"],
    
    "Salary": [25000, 30000, np.nan, 28000, 35000,
               32000, 40000, np.nan, 29000, 31000]
}

df = pd.DataFrame(data)

print(df)

print(df.info())

print(df.describe())

print(df[["Name", "City"]])

df = df.drop("City", axis=1)

mean_salary = df["Salary"].mean()

df["Salary"] = df["Salary"].fillna(mean_salary)

df = df.drop_duplicates()

print(df)
```
