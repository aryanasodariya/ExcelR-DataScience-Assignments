```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load Dataset

df = pd.read_csv("sales_data_with_discounts (1).csv")

print(df.head())
```

```python
# Identify Numerical Columns

numerical_columns = df.select_dtypes(include=[np.number]).columns

print(numerical_columns)
```

```python
# Mean

print(df[numerical_columns].mean())
```

```python
# Median

print(df[numerical_columns].median())
```

```python
# Mode

print(df[numerical_columns].mode())
```

```python
# Standard Deviation

print(df[numerical_columns].std())
```

```python
# Histogram for Sales

plt.figure(figsize=(6,4))
plt.hist(df['Total Sales Value'], bins=10)
plt.title("Histogram of Sales")
plt.xlabel("Total Sales Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Histogram for Volume

plt.figure(figsize=(6,4))
plt.hist(df['Volume'], bins=10)
plt.title("Histogram of Volume")
plt.xlabel("Volume")
plt.ylabel("Frequency")
plt.show()
```

```python
# Histogram for Avg Price

plt.figure(figsize=(6,4))
plt.hist(df['Avg Price'], bins=10)
plt.title("Histogram of Avg Price")
plt.xlabel("Avg Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# Histogram for Net Sales Value

plt.figure(figsize=(6,4))
plt.hist(df['Net Sales Value'], bins=10)
plt.title("Histogram of Net Sales Value")
plt.xlabel("Net Sales Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Histogram for Discount Amount

plt.figure(figsize=(6,4))
plt.hist(df['Discount Amount'], bins=10)
plt.title("Histogram of Discount Amount")
plt.xlabel("Discount Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
# Histogram for Discount Rate (%)

plt.figure(figsize=(6,4))
plt.hist(df['Discount Rate (%)'], bins=10)
plt.title("Histogram of Discount Rate (%)")
plt.xlabel("Discount Rate (%)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Box Plot for Net Sales Value

plt.figure(figsize=(6,4))
plt.boxplot(df['Net Sales Value'].dropna(), vert=False)
plt.title("Boxplot of Net Sales Value")
plt.xlabel("Net Sales Value")
plt.show()
```

```python
# Box Plot for Discount Amount

plt.figure(figsize=(6,4))
plt.boxplot(df['Discount Amount'].dropna(), vert=False)
plt.title("Boxplot of Discount Amount")
plt.xlabel("Discount Amount")
plt.show()
```

```python
# Box Plot for Discount Rate (%)

plt.figure(figsize=(6,4))
plt.boxplot(df['Discount Rate (%)'].dropna(), vert=False)
plt.title("Boxplot of Discount Rate (%)")
plt.xlabel("Discount Rate (%)")
plt.show()
```

```python
# Box Plot for Total Sales Value

plt.figure(figsize=(6,4))
plt.boxplot(df['Total Sales Value'].dropna(), vert=False)
plt.title("Boxplot of Total Sales Value")
plt.xlabel("Total Sales Value")
plt.show()
```

```python
# Box Plot for Avg Price

plt.figure(figsize=(6,4))
plt.boxplot(df['Avg Price'].dropna(), vert=False)
plt.title("Boxplot of Avg Price")
plt.xlabel("Avg Price")
plt.show()
```

```python
# Box Plot for Volume

plt.figure(figsize=(6,4))
plt.boxplot(df['Volume'].dropna(), vert=False)
plt.title("Boxplot of Volume")
plt.xlabel("Volume")
plt.show()
```

```python
# Identify Categorical Columns

categorical_columns = df.select_dtypes(include=['object']).columns

print(categorical_columns)
```

```python
# Bar Chart for Volume

df['Volume'].value_counts().plot(kind='bar', figsize=(6,4))
plt.title("Bar Chart of Volume")
plt.xlabel("Volume")
plt.ylabel("Count")
plt.show()
```

```python
# Bar Chart for Net Sales Value

plt.figure(figsize=(8,4))

df['Net Sales Value'].head(20).plot(kind='bar')

plt.title("Bar Chart of Net Sales Value")
plt.xlabel("Index")
plt.ylabel("Net Sales Value")

plt.show()
```

```python
# Bar Chart for Discount Amount

plt.figure(figsize=(8,4))

df['Discount Amount'].head(20).plot(kind='bar')

plt.title("Bar Chart of Discount Amount")
plt.xlabel("Index")
plt.ylabel("Discount Amount")

plt.show()
```

```python
# Bar Chart for Discount Rate (%)

plt.figure(figsize=(8,4))

df['Discount Rate (%)'].head(20).plot(kind='bar')

plt.title("Bar Chart of Discount Rate (%)")
plt.xlabel("Index")
plt.ylabel("Discount Rate (%)")

plt.show()
```

```python
# Bar Chart for Total Sales Value

plt.figure(figsize=(8,4))

df['Total Sales Value'].head(20).plot(kind='bar')

plt.title("Bar Chart of Total Sales Value")
plt.xlabel("Index")
plt.ylabel("Total Sales Value")

plt.show()
```

```python
# Bar Chart for Avg Price

df['Avg Price'].value_counts().plot(kind='bar', figsize=(6,4))
plt.title("Bar Chart of Avg Price")
plt.xlabel("Avg Price")
plt.ylabel("Count")
plt.show()
```
