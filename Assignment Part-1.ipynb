```python
# Exercise 1: Python program to check whether a number is prime or not

num = int(input("Enter a number: "))

if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print(num, "is not a Prime Number")
            break
    else:
        print(num, "is a Prime Number")
else:
    print(num, "is not a Prime Number")
```

```python
# Exercise 2: Product of Random Numbers

import random

num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

answer = int(input(f"What is the product of {num1} and {num2}? "))

correct_product = num1 * num2

if answer == correct_product:
    print("Correct Answer!")
else:
    print("Wrong Answer!")
    print("The correct product is:", correct_product)
```

```python
# Exercise 3: Squares of Even Numbers

for i in range(100, 201):
    if i % 2 == 0:
        print("Number:", i, "Square:", i ** 2)
```

```python
# Exercise 4: Hospital Billing System

room_charges = 4 * 2000
doctor_fee = 1500
lab_charges = 3 * 300
medicine_charges = 2400

total_bill = room_charges + doctor_fee + lab_charges + medicine_charges

if total_bill > 10000:
    discount = total_bill * 0.10
else:
    discount = 0

final_bill = total_bill - discount

print("Total Bill:", total_bill)
print("Discount:", discount)
print("Final Bill:", final_bill)
```
