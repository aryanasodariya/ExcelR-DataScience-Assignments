```python
# Task 1: Import Required Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Task 2: Load the Dataset

df = pd.read_csv("Cardiotocographic.csv")

print(df.head())
```

```python
# Task 3: Display Dataset Information

print(df.info())
```

```python
# Task 4: Check Missing Values

print(df.isnull().sum())
```

```python
# Task 5: Fill Missing Values

df = df.fillna(df.mean(numeric_only=True))

print(df.isnull().sum())
```

```python
# Task 6: Check Data Types

print(df.dtypes)
```

```python
# Task 7: Statistical Summary

print(df.describe())
```

```python
# Task 8: Calculate Median

print(df.median(numeric_only=True))
```

```python
# Task 9: Calculate Standard Deviation

print(df.std(numeric_only=True))
```

```python
# Task 10: Calculate Interquartile Range

Q1 = df.quantile(0.25, numeric_only=True)
Q3 = df.quantile(0.75, numeric_only=True)

IQR = Q3 - Q1

print(IQR)
```

```python
# Task 11: Histogram for LB

plt.figure(figsize=(6,4))

plt.hist(df['LB'], bins=10)

plt.title("Histogram of LB")
plt.xlabel("LB")
plt.ylabel("Frequency")

plt.show()
```

```python
# Task 12: Box Plot for LB

plt.figure(figsize=(6,4))

plt.boxplot(df['LB'].dropna(), vert=False)

plt.title("Box Plot of LB")
plt.xlabel("LB")

plt.show()
```

```python
# Task 13: Histogram for AC

plt.figure(figsize=(6,4))

plt.hist(df['AC'], bins=10)

plt.title("Histogram of AC")
plt.xlabel("AC")
plt.ylabel("Frequency")

plt.show()
```

```python
# Task 14: Box Plot for AC

plt.figure(figsize=(6,4))

plt.boxplot(df['AC'].dropna(), vert=False)

plt.title("Box Plot of AC")
plt.xlabel("AC")

plt.show()
```

```python
# Task 15: Scatter Plot

plt.figure(figsize=(6,4))

plt.scatter(df['LB'], df['AC'])

plt.title("Scatter Plot of LB vs AC")
plt.xlabel("LB")
plt.ylabel("AC")

plt.show()
```

```python
# Task 16: Pair Plot

sns.pairplot(df[['LB', 'AC', 'FM', 'UC']])

plt.show()
```

```python
# Task 17: Detect and Treat Outliers Using IQR Method

numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = np.where(df[col] < lower, lower, df[col])
    df[col] = np.where(df[col] > upper, upper, df[col])

print("Outlier Treatment Completed")
```

```python
# Task 18: Bar Chart for NSP (Categorical Variable)

plt.figure(figsize=(6,4))

df['NSP'].value_counts().plot(kind='bar')

plt.title("Bar Chart of NSP")
plt.xlabel("NSP")
plt.ylabel("Count")

plt.show()
```

```python
# Task 20: Correlation Heatmap

plt.figure(figsize=(12,8))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
# Task 21: Violin Plot for LB

plt.figure(figsize=(6,4))

sns.violinplot(x=df['LB'])

plt.title("Violin Plot of LB")

plt.show()
```

```python
# Task 22: Violin Plot for AC

plt.figure(figsize=(6,4))

sns.violinplot(x=df['AC'])

plt.title("Violin Plot of AC")

plt.show()
```

```python
# Task 23: Correlation Analysis

corr_matrix = df.corr(numeric_only=True)

print(corr_matrix)
```
